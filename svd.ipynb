```python
import numpy as np
AAT = np.array([[2, -1, 0],
              [-1, 1, -1], 
              [0, -1, 2]])

lambdas, V = np.linalg.eig(AAT)
print(lambdas)
print(V)

```

```python
A = np.array([[-1, 1],
              [0, -1], 
              [1, 1]])
U, D, VT = np.linalg.svd(A)
print('U: \n', U)
print('D: \n', D)
print('VT: \n', VT)
```

```python
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt
img = Image.open('batman.jpg')
plt.imshow(img)
```

# convert to gray scale

```python
imGray = img.convert('LA')
plt.imshow(imGray)
```

# Convert data into numpy matrix

```python
imgMat = np.array(list(imGray.getdata(band=0)), float)
imgMat.shape = (imGray.size[1], imGray.size[0])
imgMat = np.matrix(imgMat)
plt.imshow(imgMat, cmap = 'gray')
```

# Calculate SVD of the image
## As eigenvalues are arranged in descending order in diag so

```python
U, sigma, V =np.linalg.svd(imgMat)
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray')
```

## Additional singular vectors improve the image quality

```python
for i in [2,4,8,16,32,64]:
    reconstimg=np.matrix(U[:, :i])*np.diag(sigma[:i])*np.matrix(V[:i,:])
    plt.imshow(reconstimg,cmap="gray")
    title= "n=%s" %i
    plt.title(title)
    plt.show()
```
